```python
import gspread
import pandas as pd
import matplotlib.pyplot as plt
from oauth2client.service_account import ServiceAccountCredentials
```

```python
scope = ['https://spreadsheets.google.com/feeds', 'https://www.googleapis.com/auth/drive']
```

```python
def createcredentials(stringname, sheetname):
    credentials = ServiceAccountCredentials.from_json_keyfile_name(stringname, scope)
    gc = gspread.authorize(credentials)
    wks = gc.open(sheetname).sheet1
    d = wks.get_all_records()
    print(type(d))
    df = pd.DataFrame(d)
    print(df)
    return df
```

```python
df = createcredentials('/home/nidhi/Programming Files/Python Projects/newapilibrary-367ec85bcf11.json', 'testapisheet')
```

```python
bleh_price = df[['score1', 'score2']].astype(float)
bleh_price.plot(figsize=(20,10), title = 'bleh')
```
